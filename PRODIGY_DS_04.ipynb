```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

# Sample data
data = {
    'date': ['2024-05-01', '2024-05-02', '2024-05-03', '2024-05-04', '2024-05-05', '2024-05-06', '2024-05-07', '2024-05-08', '2024-05-09'],
    'text': [
        'I love this product!',
        'This product is terrible.',
        'Neutral opinion about the product.',
        'This product is amazing!',
        'I dislike this product.',
        'Great experience with this brand!',
        'I had a horrible experience with this brand.',
        'Mixed feelings about this brand.',
        'This brand consistently delivers high-quality products.'
    ]
}

df = pd.DataFrame(data)

# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Sentiment analysis
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()

# Function to get sentiment score
def get_sentiment_score(text):
    return sid.polarity_scores(text)['compound']

# Apply sentiment analysis to each text
df['sentiment_score'] = df['text'].apply(get_sentiment_score)

# Plot sentiment over time
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['sentiment_score'], marker='o', color='skyblue')
plt.xlabel('Date')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Over Time')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
